```python
from transformers import MarianMTModel, MarianTokenizer

# 사용할 모델의 이름
model_name = 'Helsinki-NLP/opus-mt-ko-en'

# 1. 토크나이저 불러오기
tokenizer = MarianTokenizer.from_pretrained(model_name)

# 2. 모델 불러오기
model = MarianMTModel.from_pretrained(model_name)

print("✅ 모델과 토크나이저를 성공적으로 불러왔습니다.")
```

```python
# 번역할 한국어 문장
korean_text = "안녕하세요, 미국의 수도는 워싱턴입니다."

# 1. 문장을 토큰화하여 모델 입력 형태로 변환
batch = tokenizer([korean_text], return_tensors="pt", padding=True)

# 2. 모델을 사용하여 번역 실행
gen = model.generate(**batch)

# 3. 번역된 결과를 다시 텍스트로 변환
english_translation = tokenizer.batch_decode(gen, skip_special_tokens=True)[0]


# 결과 출력
print("🇰🇷 한국어 원문:")
print(korean_text)
print("-" * 30)
print("🇺🇸 영어 번역 결과:")
print(english_translation)

# 다른 문장으로 테스트
print("\n" + "="*30 + "\n")

korean_text_2 = "충북 진천의 날씨 좋습니다."
batch_2 = tokenizer([korean_text_2], return_tensors="pt", padding=True)
gen_2 = model.generate(**batch_2)
english_translation_2 = tokenizer.batch_decode(gen_2, skip_special_tokens=True)[0]

print("🇰🇷 한국어 원문:")
print(korean_text_2)
print("-" * 30)
print("🇺🇸 영어 번역 결과:")
print(english_translation_2)
```
